```python
# ============================================================================
# ADDIS ABABA RAINFALL ANALYSIS — All Figures
# Google Colab / Jupyter Notebook compatible

# USAGE IN GOOGLE COLAB:
# 1. Upload these two files to your Colab session (Files panel on left):
#    - Addis_Ababa_Rainfall_CORRECTED.xlsx
#    - 24hrdepth.csv
# 2. Run the INSTALL cell once (first cell below)
# 3. Run remaining cells in order

# CORRECTIONS APPLIED IN THIS SCRIPT vs. thesis text:
# - Pettitt K = 168 (NOT 41), p = 0.2145 (NOT 1.75)
# - ERA values = Table 5.19 from uploaded CSV ONLY
# ============================================================================

# ── CELL 1: Install / Import ──────────────────────────────────────────────
# Run this cell first in Colab
# !pip install openpyxl scipy # uncomment if needed

import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
import os

print(f"pandas {pd.__version__} | numpy {np.__version__} | matplotlib {matplotlib.__version__}")

# ── CELL 2: File paths — EDIT THESE if your files are in a different location ─
EXCEL_FILE = 'Addis_Ababa_Rainfall_CORRECTED.xlsx'
ERA_CSV = '24hrdepth.csv'

# Output folder for saved figures
FIG_DIR = 'thesis_figures'
os.makedirs(FIG_DIR, exist_ok=True)

# Verify files exist
for f in [EXCEL_FILE, ERA_CSV]:
    if os.path.exists(f):
        print(f"✓ Found: {f}")
    else:
        print(f"✗ NOT FOUND: {f} ← upload this file to Colab first")

# ── CELL 3: Helper functions ──────────────────────────────────────────────
xl = pd.ExcelFile(EXCEL_FILE)

def read_sheet(name):
    """Read a sheet, auto-detecting the header row."""
    df = pd.read_excel(xl, sheet_name=name, header=None)
    header_row = None
    for i, row in df.iterrows():
        val = str(row.iloc[0]).strip()
        if val in ('Year', 'Statistic', 'Rank', 'Bin Centre (mm)', 'Month',
                   'T (yr)', 'Return Period T (yr)', 'Source', 'Metric'):
            header_row = i
            break
    if header_row is None:
        header_row = 2
    df.columns = [str(c).strip() for c in df.iloc[header_row]]
    df = df.iloc[header_row + 1:].reset_index(drop=True)
    return df

def to_num(df, cols):
    """Convert specified columns to numeric."""
    for c in cols:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors='coerce')
    return df

# ERA Table 5.19 — only valid ERA source
era = pd.read_csv(ERA_CSV)
era.columns = ['T_yr', 'R24_mm']
print("\nERA Table 5.19:")
print(era.to_string(index=False))

# Colour palette
B  = '#1f77b4'  # Blue = GT
O  = '#ff7f0e'  # Orange = NOAA
G  = '#2ca02c'  # Green = BOLE
R  = '#d62728'  # Red = trend/bias/ERA
P  = '#9467bd'  # Purple = GEV
GR = '#7f7f7f'  # Grey = reference

plt.rcParams.update({
    'font.family': 'DejaVu Sans',
    'font.size': 10,
    'axes.spines.top': False,
    'axes.spines.right': False,
    'axes.grid': True,
    'grid.alpha': 0.3,
    'figure.dpi': 120,
})

def savefig(name, fig=None):
    path = os.path.join(FIG_DIR, f'{name}.png')
    plt.tight_layout(pad=2)
    plt.savefig(path, bbox_inches='tight', dpi=150)
    plt.show()
    print(f"Saved → {path}")

print("\n✓ Setup complete. All helper functions loaded.")

# =============================================================================
# ── CELL 4: FIGURE 1 — GT Annual Totals (Sheet 01) ───────────────────────────
# Source: 01_GT_Annual_Totals
# Calculation: Bar chart coloured by above/below mean; mean ± 1SD reference lines
# =============================================================================
d1 = to_num(read_sheet('01_GT_Annual_Totals'), ['Year', 'Annual Total (mm)'])
d1 = d1.dropna(subset=['Year', 'Annual Total (mm)'])
mean1 = d1['Annual Total (mm)'].mean()
sd1 = d1['Annual Total (mm)'].std(ddof=1)
cv1 = sd1 / mean1

print(f"Sheet 01 | n={len(d1)} | Mean={mean1:.2f} mm | SD={sd1:.2f} mm | CV={cv1:.4f}")
print(f" Min={d1['Annual Total (mm)'].min()} mm ({int(d1.loc[d1['Annual Total (mm)'].idxmin(),'Year'])})")
print(f" Max={d1['Annual Total (mm)'].max()} mm ({int(d1.loc[d1['Annual Total (mm)'].idxmax(),'Year'])})")

above1 = d1['Annual Total (mm)'] >= mean1

fig, ax = plt.subplots(figsize=(14, 6))
ax.bar(
    d1['Year'], d1['Annual Total (mm)'],
    color=[B if a else O for a in above1],
    alpha=0.82, width=0.9, edgecolor='white', lw=0.3
)
ax.axhline(mean1, color=R, lw=2, ls='--',
           label=f'Long-term mean = {mean1:.1f} mm/yr')
ax.axhline(
    mean1 + sd1, color=GR, lw=1.2, ls=':', alpha=0.8,
    label=f'Mean ± 1 SD ({mean1-sd1:.0f}–{mean1+sd1:.0f} mm)'
)
ax.axhline(mean1 - sd1, color=GR, lw=1.2, ls=':', alpha=0.8)

# Annotate exceptionally wet years (≥1700 mm)
for _, row in d1[d1['Annual Total (mm)'] >= 1700].iterrows():
    ax.annotate(
        f"{int(row['Year'])}\n{row['Annual Total (mm)']:.0f} mm",
        xy=(row['Year'], row['Annual Total (mm)']),
        xytext=(0, 6), textcoords='offset points',
        ha='center', fontsize=8, color='navy', fontweight='bold'
    )

p_above = mpatches.Patch(color=B, alpha=0.82,
                         label=f'Above mean (n={above1.sum()})')
p_below = mpatches.Patch(color=O, alpha=0.82,
                         label=f'Below mean (n={(~above1).sum()})')
ax.legend(handles=[p_above, p_below] + ax.lines[:2],
          fontsize=9, loc='upper left')
ax.set_xlim(1896, 1942)
ax.set_xlabel('Year')
ax.set_ylabel('Annual Rainfall Total (mm)')
ax.set_title(
    f'Figure 1 – GT Annual Rainfall Totals: Addis Abeba Observatory 1898–1940\n'
    f'n=42 | Mean={mean1:.1f} mm | SD={sd1:.1f} mm | CV={cv1:.3f}',
    fontweight='bold'
)
savefig('Fig01_GT_Annual_Totals')

# =============================================================================
# ── CELL 5: FIGURE 2 — NOAA vs GT Annual Max Daily (Sheet 02) ────────────────
# Source: 02_NOAA_vs_GT_Daily
# Calculation: Time series + side-by-side bar for 9 overlap years; GT/NOAA ratio
# =============================================================================
d2 = to_num(
    read_sheet('02_NOAA_vs_GT_Daily'),
    ['Year', 'NOAA Max Daily (mm)', 'GT Max Daily (mm)', 'GT/NOAA Ratio']
)
dn = d2.dropna(subset=['NOAA Max Daily (mm)'])
dg = d2.dropna(subset=['GT Max Daily (mm)'])
ovl = d2.dropna(subset=['GT Max Daily (mm)', 'NOAA Max Daily (mm)'])

print(f"\nSheet 02 | NOAA n={len(dn)} mean={dn['NOAA Max Daily (mm)'].mean():.2f} mm")
print(f" | GT n={len(dg)} mean={dg['GT Max Daily (mm)'].mean():.2f} mm")
print(
    f" | 9 overlap years, GT/NOAA ratio: "
    f"{(ovl['GT Max Daily (mm)']/ovl['NOAA Max Daily (mm)']).mean():.2f}×"
)

fig, axes = plt.subplots(2, 1, figsize=(14, 11))

ax = axes[0]
ax.plot(
    dn['Year'], dn['NOAA Max Daily (mm)'],
    color=O, lw=1.8, marker='o', ms=4,
    label=f'NOAA 20CRv3 (n={len(dn)}, mean={dn["NOAA Max Daily (mm)"].mean():.1f} mm)'
)
ax.plot(
    dg['Year'], dg['GT Max Daily (mm)'],
    color=B, lw=2.5, marker='s', ms=7,
    label=f'GT Daily (n={len(dg)}, mean={dg["GT Max Daily (mm)"].mean():.1f} mm)'
)
ax.set_xlabel('Year')
ax.set_ylabel('Annual Max Daily Rainfall (mm)')
ax.set_title(
    f'(a) Annual Maximum Daily Rainfall Time Series: NOAA 20CRv3 vs Ground Truth\n'
    f'Only 9 overlap years have GT daily data: 1923–1926, 1933, 1937–1940',
    fontweight='bold'
)
ax.legend(fontsize=9)

ax2 = axes[1]
ratio_mean = (ovl['GT Max Daily (mm)'] / ovl['NOAA Max Daily (mm)']).mean()
xr = np.arange(len(ovl))
ax2.bar(
    xr - 0.2, ovl['GT Max Daily (mm)'], 0.38,
    color=B, alpha=0.85, label='GT Max Daily'
)
ax2.bar(
    xr + 0.2, ovl['NOAA Max Daily (mm)'], 0.38,
    color=O, alpha=0.85, label='NOAA Max Daily'
)
ax2.set_xticks(xr)
ax2.set_xticklabels(ovl['Year'].astype(int), rotation=45)
ax2.set_xlabel('Year (9 overlap years)')
ax2.set_ylabel('Max Daily Rainfall (mm)')
ax2.set_title(
    f'(b) Side-by-Side Comparison (9 Overlap Years)\n'
    f'GT/NOAA mean ratio = {ratio_mean:.2f}× | NOAA underestimates by ~{(1-1/ratio_mean)*100:.0f}%',
    fontweight='bold'
)
ax2.legend(fontsize=9)

fig.suptitle(
    'Figure 2 – NOAA 20CRv3 vs Ground Truth: Annual Max Daily Rainfall',
    fontsize=12, fontweight='bold', y=1.01
)
savefig('Fig02_NOAA_vs_GT_Daily')

# =============================================================================
# ── CELL 6: FIGURE 3 — BOLE Annual Max Daily (Sheet 03) ──────────────────────
# Source: 03_BOLE_Annual_Max
# Calculation: Bar coloured by above/below mean; 7 excluded years noted
# =============================================================================
d3 = to_num(read_sheet('03_BOLE_Annual_Max'),
            ['Year', 'Ann Max Daily (mm)'])
d3 = d3.dropna(subset=['Year', 'Ann Max Daily (mm)'])
mean3 = d3['Ann Max Daily (mm)'].mean()
sd3 = d3['Ann Max Daily (mm)'].std(ddof=1)
cv3 = sd3 / mean3

print(
    f"\nSheet 03 | n={len(d3)} | Mean={mean3:.2f} mm | SD={sd3:.2f} mm | CV={cv3:.4f}"
)
print(
    " Excluded years: 1960, 1961, 1999, 2000, 2002, 2003, 2004 "
    "(< 30 non-zero daily obs)"
)

above3 = d3['Ann Max Daily (mm)'] >= mean3

fig, ax = plt.subplots(figsize=(14, 6))
ax.bar(
    d3['Year'], d3['Ann Max Daily (mm)'],
    color=[G if a else GR for a in above3],
    alpha=0.82, width=0.9, edgecolor='white', lw=0.3
)
ax.axhline(mean3, color=R, lw=2, ls='--',
           label=f'Mean = {mean3:.1f} mm')
ax.axhline(
    mean3 + sd3, color='brown', lw=1.2, ls=':', alpha=0.8,
    label=f'Mean ± 1 SD ({mean3-sd3:.1f}–{mean3+sd3:.1f} mm)'
)
ax.axhline(mean3 - sd3, color='brown', lw=1.2, ls=':', alpha=0.8)

p3a = mpatches.Patch(color=G,  alpha=0.82,
                     label=f'Above mean (n={above3.sum()})')
p3b = mpatches.Patch(color=GR, alpha=0.82,
                     label=f'Below mean (n={(~above3).sum()})')
ax.legend(handles=[p3a, p3b] + ax.lines[:2], fontsize=9)

ax.set_xlim(1955, 2022)
ax.set_xlabel('Year')
ax.set_ylabel('Annual Max Daily Rainfall (mm)')
ax.set_title(
    f'Figure 3 – BOLE GHCND Station Annual Max Daily Rainfall 1957–2020\n'
    f'n={len(d3)} valid years (7 excluded) | Mean={mean3:.1f} mm | SD={sd3:.1f} mm | CV={cv3:.3f}',
    fontweight='bold'
)
savefig('Fig03_BOLE_Annual_Max')

# =============================================================================
# ── CELL 7: FIGURE 4 — Boxplot & Descriptive Stats (Sheet 04) ────────────────
# Source: 04_Boxplot_Stats (table) + raw data from Sheets 03, 15, 02
# =============================================================================
d4 = read_sheet('04_Boxplot_Stats')
bole_r = d3['Ann Max Daily (mm)']
gt_r = to_num(read_sheet('15_GT_Prob_Plot'),
              ['Observed (mm)']).dropna(subset=['Observed (mm)'])['Observed (mm)']
noaa_r = to_num(read_sheet('02_NOAA_vs_GT_Daily'),
                ['NOAA Max Daily (mm)']).dropna(subset=['NOAA Max Daily (mm)'])['NOAA Max Daily (mm)']

print(
    f"\nSheet 04 | BOLE n={len(bole_r)} | GT n={len(gt_r)} | NOAA n={len(noaa_r)}"
)
print(d4.head(9).to_string())

fig, axes = plt.subplots(1, 2, figsize=(16, 7))

ax = axes[0]
bp = ax.boxplot(
    [bole_r, gt_r, noaa_r],
    labels=[
        f'BOLE\n(n={len(bole_r)})',
        f'GT Daily\n(n={len(gt_r)})',
        f'NOAA Daily\n(n={len(noaa_r)})'
    ],
    patch_artist=True,
    medianprops=dict(color=R, lw=2.5),
    widths=0.5
)
for patch, c in zip(bp['boxes'], [G, B, O]):
    patch.set_facecolor(c)
    patch.set_alpha(0.55)
ax.set_ylabel('Annual Max Daily Rainfall (mm)')
ax.set_title('(a) Boxplot: All Sources', fontweight='bold')

ax2 = axes[1]
ax2.axis('off')
tbl_data = [
    [str(r.iloc[0]), str(r.iloc[1]), str(r.iloc[2]), str(r.iloc[3])]
    for _, r in d4.iterrows()
]
tbl = ax2.table(
    cellText=tbl_data,
    colLabels=['Statistic', 'BOLE (n=54)',
               'GT Daily (n=9)', 'NOAA Daily (n=44)'],
    loc='center', cellLoc='center'
)
tbl.auto_set_font_size(False)
tbl.set_fontsize(9.5)
tbl.scale(1.1, 1.9)
for (r, c), cell in tbl.get_celld().items():
    if r == 0:
        cell.set_facecolor('#2E75B6')
        cell.set_text_props(color='white', weight='bold')
    elif r % 2 == 0:
        cell.set_facecolor('#f2f2f2')
ax2.set_title('(b) Descriptive Statistics Table', fontweight='bold')

fig.suptitle(
    'Figure 4 – Boxplot & Stats: Annual Max Daily Rainfall',
    fontsize=12, fontweight='bold', y=1.01
)
savefig('Fig04_Boxplot_Stats')

# =============================================================================
# ── CELL 8: FIGURE 5 — GT Histogram + Normal PDF (Sheet 05) ──────────────────
# Source: 05_GT_Histogram
# Formula: f(x) = (1/σ√2π) × exp[−(x−μ)²/2σ²]
# =============================================================================
d5 = to_num(
    read_sheet('05_GT_Histogram'),
    ['Bin Centre (mm)', 'Count', 'Density', 'Normal PDF']
)
d5 = d5.dropna(subset=['Bin Centre (mm)', 'Count'])
bw5 = float(d5['Bin Centre (mm)'].diff().median()) * 0.9

print(
    f"\nSheet 05 | {len(d5)} bins | μ={mean1:.2f} mm | σ={sd1:.2f} mm"
)
print(
    " Normal PDF formula: f(x) = (1/(σ√2π)) × exp[−(x−μ)²/(2σ²)]"
)

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(
    d5['Bin Centre (mm)'], d5['Density'], width=bw5,
    color=B, alpha=0.65, edgecolor='white',
    label='Observed frequency density'
)
ax.plot(
    d5['Bin Centre (mm)'], d5['Normal PDF'],
    color=R, lw=2.5, marker='o', ms=4,
    label=f'Normal PDF | μ={mean1:.1f} mm, σ={sd1:.1f} mm'
)
for _, row in d5.iterrows():
    if row['Count'] > 0:
        ax.text(
            row['Bin Centre (mm)'], row['Density'] + 0.00005,
            f"n={int(row['Count'])}",
            ha='center', fontsize=7.5, color='navy'
        )
ax.set_xlabel('Annual Total (mm)')
ax.set_ylabel('Density')
ax.set_title(
    f'Figure 5 – GT Annual Totals: Histogram + Normal PDF\n'
    f'n=42 | μ={mean1:.1f} mm | σ={sd1:.1f} mm | Bin width ≈ 96.8 mm\n'
    f'Normal PDF: f(x) = (1/σ√2π) × exp[−(x−μ)²/2σ²]',
    fontweight='bold'
)
ax.legend(fontsize=10)
savefig('Fig05_GT_Histogram')

# =============================================================================
# ── CELL 9: FIGURE 6 — BOLE Histogram + Gumbel EV1 PDF (Sheet 06) ────────────
# Source: 06_BOLE_Histogram
# Gumbel EV1 params (Method of Moments): β = σ√6/π, u = μ − 0.5772β
# PDF: f(x) = (1/β)×exp(−z)×exp(−exp(−z)) where z=(x−u)/β
# =============================================================================
d6 = to_num(
    read_sheet('06_BOLE_Histogram'),
    ['Bin Centre (mm)', 'Count', 'Density', 'Gumbel PDF', 'Normal PDF']
)
d6 = d6.dropna(subset=['Bin Centre (mm)', 'Count'])
bw6 = float(d6['Bin Centre (mm)'].diff().median()) * 0.9

# Method of Moments parameters for BOLE
beta3 = sd3 * np.sqrt(6) / np.pi
u3 = mean3 - 0.5772 * beta3

print(f"\nSheet 06 | {len(d6)} bins")
print(f" Gumbel MOM: β = σ√6/π = {sd3:.4f}×{np.sqrt(6)/np.pi:.4f} = {beta3:.3f} mm")
print(f" u = μ − 0.5772β = {mean3:.4f} − {0.5772*beta3:.4f} = {u3:.3f} mm")
print(" Sheet 24 values: u=39.612 mm, β=14.028 mm ← use these (confirmed)")

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(
    d6['Bin Centre (mm)'], d6['Density'], width=bw6,
    color=G, alpha=0.65, edgecolor='white',
    label='Observed frequency density'
)
ax.plot(
    d6['Bin Centre (mm)'], d6['Gumbel PDF'],
    color=R, lw=2.5, marker='o', ms=4,
    label=f'Gumbel EV1 PDF | u={u3:.2f} mm, β={beta3:.2f} mm'
)
ax.plot(
    d6['Bin Centre (mm)'], d6['Normal PDF'],
    color=GR, lw=1.8, ls='--', label='Normal PDF (reference)'
)
for _, row in d6.iterrows():
    if row['Count'] > 0:
        ax.text(
            row['Bin Centre (mm)'], row['Density'] + 0.0003,
            f"n={int(row['Count'])}",
            ha='center', fontsize=7.5, color='darkgreen'
        )
ax.set_xlabel('Annual Max Daily Rainfall (mm)')
ax.set_ylabel('Density')
ax.set_title(
    f'Figure 6 – BOLE Histogram + Gumbel EV1 PDF\n'
    f'n=55 | μ={mean3:.2f} mm | σ={sd3:.2f} mm\n'
    f'Gumbel EV1: f(x)=(1/β)·exp(−z)·exp(−e⁻ᶻ) z=(x−u)/β | u={u3:.2f}, β={beta3:.2f}',
    fontweight='bold'
)
ax.legend(fontsize=10)
savefig('Fig06_BOLE_Histogram')

# =============================================================================
# ── CELL 10: FIGURE 7 — GT Annual Total Trend (Sheet 07) ─────────────────────
# Source: 07_GT_Trend
# OLS: scipy.stats.linregress | Mann–Kendall τ and p from Sheet annotation
# =============================================================================
d7 = to_num(read_sheet('07_GT_Trend'),
            ['Year', 'Annual Total (mm)', 'OLS Trend (mm)'])
d7 = d7.dropna(subset=['Year', 'Annual Total (mm)'])
sl7, ic7, r7, p7, se7 = stats.linregress(
    d7['Year'], d7['Annual Total (mm)']
)

print(
    f"\nSheet 07 | OLS slope={sl7:.3f} mm/yr ({sl7*10:.2f} mm/decade) | "
    f"r={r7:.4f} | p={p7:.4f}"
)
print(" Mann–Kendall τ=−0.076, p=0.481 (from sheet annotation)")
print(" Conclusion: NO significant trend at any conventional significance level")

fig, axes = plt.subplots(2, 1, figsize=(14, 11))

ax = axes[0]
ax.bar(
    d7['Year'], d7['Annual Total (mm)'],
    color=B, alpha=0.55, width=0.9
)
ax.plot(
    d7['Year'],
    d7['OLS Trend (mm)'] if 'OLS Trend (mm)' in d7.columns
    else ic7 + sl7 * d7['Year'],
    color=R, lw=2.5, ls='--',
    label=(
        f'OLS: y = {ic7:.0f} + {sl7:.3f}×Year '
        f'(slope={sl7:.3f} mm/yr, p={p7:.3f})'
    )
)
ax.axhline(
    mean1, color=GR, lw=1.5, ls=':',
    label=f'Long-term mean = {mean1:.1f} mm'
)
ax.set_xlabel('Year')
ax.set_ylabel('Annual Total (mm)')
ax.set_title(
    f'(a) GT Annual Totals with OLS Trend Line\n'
    f'Mann–Kendall τ=−0.076, p=0.481 | '
    f'OLS slope={sl7:.3f} mm/yr, p={p7:.3f} → NO significant trend',
    fontweight='bold'
)
ax.legend(fontsize=8.5)

ax2 = axes[1]
trend_vals = ic7 + sl7 * d7['Year']
deviation = d7['Annual Total (mm)'] - trend_vals
ax2.bar(
    d7['Year'], deviation,
    color=[B if v >= 0 else O for v in deviation],
    alpha=0.75, width=0.9
)
ax2.axhline(0, color='black', lw=1)
ax2.set_xlabel('Year')
ax2.set_ylabel('Deviation from OLS Trend (mm)')
ax2.set_title(
    f'(b) Deviation from OLS Trend | Max above: +{deviation.max():.0f} mm | '
    f'Max below: {deviation.min():.0f} mm',
    fontweight='bold'
)

fig.suptitle(
    'Figure 7 – GT Annual Total Trend Analysis | 1898–1940',
    fontsize=12, fontweight='bold', y=1.01
)
savefig('Fig07_GT_Trend')

# =============================================================================
# ── CELL 11: FIGURE 8 — BOLE Annual Max Trend (Sheet 08) ─────────────────────
# Source: 08_BOLE_Trend
# Same procedure as Figure 7
# =============================================================================
d8 = to_num(read_sheet('08_BOLE_Trend'),
            ['Year', 'Ann Max Daily (mm)', 'OLS Trend (mm)'])
d8 = d8.dropna(subset=['Year', 'Ann Max Daily (mm)'])
sl8, ic8, r8, p8, se8 = stats.linregress(
    d8['Year'], d8['Ann Max Daily (mm)']
)

print(
    f"\nSheet 08 | OLS slope={sl8:.4f} mm/yr | r={r8:.4f} | p={p8:.4f}"
)
print(
    " Mann–Kendall τ=−0.005, p=0.958 | Conclusion: NO significant trend"
)

fig, axes = plt.subplots(2, 1, figsize=(14, 11))

ax = axes[0]
ax.bar(
    d8['Year'], d8['Ann Max Daily (mm)'],
    color=G, alpha=0.55, width=0.9
)
ax.plot(
    d8['Year'],
    d8['OLS Trend (mm)'] if 'OLS Trend (mm)' in d8.columns
    else ic8 + sl8 * d8['Year'],
    color=R, lw=2.5, ls='--',
    label=(
        f'OLS: y = {ic8:.1f} + {sl8:.4f}×Year '
        f'(slope={sl8:.4f} mm/yr, p={p8:.3f})'
    )
)
ax.axhline(
    mean3, color=GR, lw=1.5, ls=':',
    label=f'Mean = {mean3:.1f} mm'
)
ax.set_xlabel('Year')
ax.set_ylabel('Annual Max Daily Rainfall (mm)')
ax.set_title(
    f'(a) BOLE Annual Max Daily with OLS Trend Line\n'
    f'Mann–Kendall τ=−0.005, p=0.958 | '
    f'OLS slope={sl8:.4f} mm/yr, p={p8:.3f} → NO significant trend',
    fontweight='bold'
)
ax.legend(fontsize=8.5)

ax2 = axes[1]
trend_vals8 = ic8 + sl8 * d8['Year']
deviation8 = d8['Ann Max Daily (mm)'] - trend_vals8
ax2.bar(
    d8['Year'], deviation8,
    color=[G if v >= 0 else GR for v in deviation8],
    alpha=0.75, width=0.9
)
ax2.axhline(0, color='black', lw=1)
ax2.set_xlabel('Year')
ax2.set_ylabel('Deviation from OLS Trend (mm)')
ax2.set_title(
    f'(b) Deviation from Trend | Max above: +{deviation8.max():.1f} mm | '
    f'Max below: {deviation8.min():.1f} mm',
    fontweight='bold'
)

fig.suptitle(
    'Figure 8 – BOLE Annual Max Daily Trend Analysis | 1957–2020',
    fontsize=12, fontweight='bold', y=1.01
)
savefig('Fig08_BOLE_Trend')

# =============================================================================
# ── CELL 12: FIGURE 9 — Pettitt Homogeneity Test (Sheet 09) ──────────────────
# Source: 09_Pettitt_Test
# =============================================================================
d9 = to_num(
    read_sheet('09_Pettitt_Test'),
    ['Year', 'Annual Total (mm)', 'U Statistic', '|U| Abs']
)
d9 = d9.dropna(subset=['Year', 'U Statistic'])

# Corrected values from sheet summary rows
K_stat = 168
p_val = 2 * np.exp(-6 * K_stat**2 / (42**3 + 42**2))  # formula
cp_year = 1930
n_val = 42

print("\nSheet 09 — Pettitt Test CORRECTED values:")
print(f" K statistic = {K_stat}")
print(f" p-value = {p_val:.4f}")
print(
    f" Formula: p = 2×exp(−6×K²/(n³+n²)) = "
    f"2×exp(−6×{K_stat}²/({n_val}³+{n_val}²))"
)
print(f" Change-point candidate year = {cp_year} (non-significant)")
print(
    f" Max |U| in series = {d9['|U| Abs'].max()} at year "
    f"{int(d9.loc[d9['|U| Abs'].idxmax(), 'Year'])}"
)
print(
    f" Conclusion: Series is HOMOGENEOUS (p={p_val:.4f} > 0.05)"
)

fig, axes = plt.subplots(2, 1, figsize=(14, 11))

ax = axes[0]
ax.fill_between(
    d9['Year'], d9['Annual Total (mm)'],
    alpha=0.25, color=B
)
ax.plot(d9['Year'], d9['Annual Total (mm)'],
        color=B, lw=1.5)
ax.axvline(
    cp_year, color=R, lw=2, ls='--',
    label=f'Candidate change-point: {cp_year} (non-significant)'
)
ax.axhline(
    mean1, color=GR, lw=1.5, ls=':',
    label=f'Mean = {mean1:.1f} mm'
)
ax.set_xlabel('Year')
ax.set_ylabel('Annual Total (mm)')
ax.set_title(
    f'(a) GT Annual Totals with Candidate Change-Point Year\n'
    f'Change-point candidate = {cp_year} | K = {K_stat} | '
    f'p = {p_val:.4f} → HOMOGENEOUS (p > 0.05)',
    fontweight='bold'
)
ax.legend(fontsize=9)

ax2 = axes[1]
ax2.plot(
    d9['Year'], d9['|U| Abs'],
    color=R, lw=2.5, marker='o', ms=4,
    label='|U_t| values'
)
max_u_year = int(d9.loc[d9['|U| Abs'].idxmax(), 'Year'])
max_u_val = d9['|U| Abs'].max()



ax2.axhline(
    K_stat, color='purple', lw=2, ls='--',
    label=(
        f'K = {K_stat} (global max of ALL |U_t|, '
        f'at year {cp_year})'
    )
)
ax2.set_xlabel('Year')
ax2.set_ylabel('|U_t| Pettitt Statistic')
ax2.set_title(
    f'(b) |U_t| Series | K = max|U_t| = {K_stat} | p = {p_val:.4f}',
    fontweight='bold'
)
ax2.legend(fontsize=9)

fig.suptitle(
    f'Figure 9 – Pettitt Homogeneity Test | n=42 | K={K_stat} | p={p_val:.4f} | Homogeneous',
    fontsize=12, fontweight='bold', y=1.01
)
savefig('Fig09_Pettitt')

# =============================================================================
# ── CELL 13: FIGURE 10 — GT vs NOAA Annual Total Time Series (Sheet 10) ──────
# Source: 10_GT_vs_NOAA_Series
# =============================================================================
d10 = to_num(
    read_sheet('10_GT_vs_NOAA_Series'),
    ['Year', 'GT (mm)', 'NOAA (mm)', 'Bias (mm)', 'Rel Bias (%)']
)
d10 = d10.dropna(subset=['Year', 'GT (mm)'])
both10 = d10.dropna(subset=['GT (mm)', 'NOAA (mm)'])
r10, p10 = stats.pearsonr(
    both10['GT (mm)'], both10['NOAA (mm)']
)
mean_bias10 = (both10['NOAA (mm)'] - both10['GT (mm)']).mean()

print(
    f"\nSheet 10 | r={r10:.3f} | p={p10:.4f} | "
    f"Mean bias={mean_bias10:.1f} mm "
    f"({mean_bias10/both10['GT (mm)'].mean()*100:.1f}%)"
)

fig, axes = plt.subplots(2, 1, figsize=(14, 11))

ax = axes[0]
ax.plot(
    d10['Year'], d10['GT (mm)'],
    color=B, lw=2.5, marker='o', ms=4,
    label=f'GT (mean={d10["GT (mm)"].mean():.0f} mm)'
)
ax.plot(
    d10['Year'], d10['NOAA (mm)'],
    color=O, lw=2, marker='s', ms=4, ls='--',
    label=f'NOAA 20CRv3 (mean={d10["NOAA (mm)"].mean():.0f} mm)'
)
ax.fill_between(
    d10['Year'], d10['GT (mm)'], d10['NOAA (mm)'],
    alpha=0.12, color=P, label='Bias region (NOAA − GT)'
)
ax.set_xlabel('Year')
ax.set_ylabel('Annual Total (mm)')
ax.set_title(
    f'(a) Annual Total Time Series | r={r10:.3f} (p={p10:.4f})\n'
    f'Mean bias (NOAA−GT) = +{mean_bias10:.1f} mm '
    f'(+{mean_bias10/both10["GT (mm)"].mean()*100:.1f}%)',
    fontweight='bold'
)
ax.legend(fontsize=9)

ax2 = axes[1]
bias_s = d10['NOAA (mm)'] - d10['GT (mm)']
ax2.bar(
    d10['Year'], bias_s,
    color=[R if v > 0 else B for v in bias_s.fillna(0)],
    alpha=0.75, width=0.9
)
ax2.axhline(0, color='black', lw=0.8)
ax2.axhline(
    mean_bias10, color=R, lw=2, ls='--',
    label=f'Mean bias = +{mean_bias10:.1f} mm'
)
ax2.set_xlabel('Year')
ax2.set_ylabel('Bias NOAA − GT (mm)')
ax2.set_title(
    f'(b) Annual Bias | Overestimate in '
    f'{(bias_s.dropna()>0).sum()} of {len(bias_s.dropna())} years',
    fontweight='bold'
)
ax2.legend(fontsize=9)

fig.suptitle(
    'Figure 10 – GT vs NOAA Annual Total',
    fontsize=12, fontweight='bold', y=1.01
)
savefig('Fig10_GT_vs_NOAA_Series')

# =============================================================================
# ── CELL 14: FIGURE 11 — Scatter GT vs NOAA (Sheet 11) ───────────────────────
# Source: 11_Scatter_GT_NOAA
# OLS regression: GT = f(NOAA)
# =============================================================================
d11 = to_num(
    read_sheet('11_Scatter_GT_NOAA'),
    ['Year', 'GT (mm)', 'NOAA (mm)']
)
d11 = d11.dropna(subset=['GT (mm)', 'NOAA (mm)'])
sl11, ic11, r11, p11, _ = stats.linregress(
    d11['NOAA (mm)'], d11['GT (mm)']
)
xs11 = np.linspace(
    d11['NOAA (mm)'].min(), d11['NOAA (mm)'].max(), 200
)

print(
    f"\nSheet 11 | OLS: GT = {sl11:.4f}×NOAA + ({ic11:.1f}) | "
    f"r={r11:.3f} | p={p11:.4f}"
)

fig, ax = plt.subplots(figsize=(9, 8))
sc = ax.scatter(
    d11['NOAA (mm)'], d11['GT (mm)'],
    c=d11['Year'], cmap='plasma', s=55, zorder=3, alpha=0.85
)
plt.colorbar(sc, ax=ax, label='Year')

lim_lo = min(
    d11['NOAA (mm)'].min(), d11['GT (mm)'].min()
) - 100
lim_hi = max(
    d11['NOAA (mm)'].max(), d11['GT (mm)'].max()
) + 100
ax.plot(
    [lim_lo, lim_hi], [lim_lo, lim_hi],
    'k--', lw=1.5, alpha=0.6, label='1:1 line'
)
ax.plot(
    xs11, sl11 * xs11 + ic11,
    color=R, lw=2.5,
    label=(
        f'OLS: GT = {sl11:.4f}×NOAA + ({ic11:.1f})\n'
        f'r={r11:.3f}, p={p11:.4f}, n={len(d11)}'
    )
)
ax.set_xlabel('NOAA 20CRv3 Annual Total (mm)')
ax.set_ylabel('GT Annual Total (mm)')
ax.set_title(
    'Figure 11 – Scatter: GT vs NOAA Annual Totals\n'
    f'GT CV={d11["GT (mm)"].std(ddof=1)/d11["GT (mm)"].mean():.3f} vs '
    f'NOAA CV={d11["NOAA (mm)"].std(ddof=1)/d11["NOAA (mm)"].mean():.3f}',
    fontweight='bold'
)
ax.legend(fontsize=9)
savefig('Fig11_Scatter')

# =============================================================================
# ── CELL 15: FIGURE 12 — Monthly Climatology (Sheet 13) ──────────────────────
# Source: 13_Monthly_Climatology
# =============================================================================
d13 = to_num(
    read_sheet('13_Monthly_Climatology'),
    ['GT Mean (mm)', 'NOAA Mean (mm)', 'Bias (mm)', 'Bias (%)']
)
d13 = d13.dropna(subset=['GT Mean (mm)'])
months13 = [
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'
]
d13['Month'] = months13[:len(d13)]

kiremt_bias = d13.iloc[5:9]['Bias (%)'].mean()
print(
    f"\nSheet 13 | GT annual sum={d13['GT Mean (mm)'].sum():.0f} mm | "
    f"NOAA={d13['NOAA Mean (mm)'].sum():.0f} mm"
)
print(
    f" Kiremt (Jun–Sep) bias mean = +{kiremt_bias:.1f}% | "
    f"Oct bias = +{d13.iloc[9]['Bias (%)']:.1f}%"
)

fig, axes = plt.subplots(2, 1, figsize=(14, 11))

ax = axes[0]
x13 = np.arange(len(d13))
ax.bar(
    x13 - 0.2, d13['GT Mean (mm)'], 0.38,
    color=B, alpha=0.85,
    label=f'GT Mean (ann sum={d13["GT Mean (mm)"].sum():.0f} mm)'
)
ax.bar(
    x13 + 0.2, d13['NOAA Mean (mm)'], 0.38,
    color=O, alpha=0.85,
    label=f'NOAA Mean (ann sum={d13["NOAA Mean (mm)"].sum():.0f} mm)'
)
ax.axvspan(
    1.5, 4.5, alpha=0.12, color='#ff9900',
    label='Belg (Mar–May)'
)
ax.axvspan(
    4.5, 8.5, alpha=0.12, color='#0099ff',
    label='Kiremt (Jun–Sep)'
)
ax.set_xticks(x13)
ax.set_xticklabels(d13['Month'])
ax.set_ylabel('Mean Monthly Rainfall (mm)')
ax.set_title(
    '(a) Mean Monthly Climatology – GT vs NOAA 20CRv3',
    fontweight='bold'
)
ax.legend(fontsize=8.5, ncol=2)

ax2 = axes[1]
ax2.bar(
    x13, d13['Bias (%)'],
    color=[R if v > 0 else B for v in d13['Bias (%)']],
    alpha=0.8
)
ax2.axhline(0, color='black', lw=0.8)
ax2.axhline(
    d13['Bias (%)'].mean(),
    color=R, lw=2, ls='--',
    label=f'Overall mean bias = +{d13["Bias (%)"].mean():.1f}%'
)
ax2.set_xticks(x13)
ax2.set_xticklabels(d13['Month'])
ax2.set_ylabel('Monthly Bias % (NOAA − GT) / GT × 100')
ax2.set_title(
    f'(b) Monthly Bias % | Kiremt mean = +{kiremt_bias:.1f}% | '
    f'Oct = +{d13.iloc[9]["Bias (%)"]:.1f}%',
    fontweight='bold'
)
ax2.legend(fontsize=9)

fig.suptitle(
    'Figure 12 – Monthly Climatology & Bias',
    fontsize=12, fontweight='bold', y=1.01
)
savefig('Fig12_Monthly_Climatology')

# =============================================================================
# ── CELL 16: FIGURE 13 — BOLE Gumbel Probability Plot (Sheet 14) ─────────────
# Gringorten plotting position: p_i = (i − 0.44) / (n + 0.12)
# Gumbel reduced variate: yT = −ln[−ln(p_i)]
# =============================================================================
d14 = to_num(
    read_sheet('14_BOLE_Prob_Plot'),
    [
        'Rank', 'Observed (mm)', 'Plot Position p',
        'Gumbel yT', 'Fitted (mm)', 'Residual (mm)'
    ]
)
d14 = d14.dropna(subset=['Observed (mm)', 'Gumbel yT'])

print(f"\nSheet 14 | BOLE Prob Plot n={len(d14)}")
print(" Gringorten: p_i = (i−0.44)/(n+0.12)")
print(" KS D=0.0668, p=0.9531 → Gumbel EV1 accepted ✓")

fig, axes = plt.subplots(1, 2, figsize=(15, 7))

ax = axes[0]
ax.scatter(
    d14['Gumbel yT'], d14['Observed (mm)'],
    color=G, s=45, zorder=3,
    label=f'BOLE observed (n={len(d14)})'
)
ax.plot(
    d14['Gumbel yT'], d14['Fitted (mm)'],
    color=R, lw=2.5,
    label='Gumbel EV1 fit\nu=39.612 mm, β=14.028 mm\nKS D=0.0668, p=0.953 ✓'
)
ax.set_xlabel('Gumbel Reduced Variate yT = −ln[−ln(1−1/T)]')
ax.set_ylabel('Annual Max Daily Rainfall (mm)')
ax.set_title(
    '(a) Gumbel Probability Plot – BOLE\n'
    'Gringorten (1963): p_i=(i−0.44)/(n+0.12)',
    fontweight='bold'
)
ax.legend(fontsize=9)

ax2 = axes[1]
ax2.bar(
    range(len(d14)), d14['Residual (mm)'],
    color=[G if v >= 0 else R for v in d14['Residual (mm)']],
    alpha=0.75
)
ax2.axhline(0, color='black', lw=1)
ax2.set_xlabel('Rank')
ax2.set_ylabel('Residual = Observed − Fitted (mm)')
ax2.set_title(
    f'(b) Residuals | Mean = {d14["Residual (mm)"].mean():.2f} mm',
    fontweight='bold'
)
fig.suptitle(
    'Figure 13 – Gumbel EV1 Probability Plot: BOLE | n=54 | KS p=0.953',
    fontsize=12, fontweight='bold', y=1.01
)
savefig('Fig13_BOLE_ProbPlot')

# =============================================================================
# ── CELL 17: FIGURE 14 — GT Daily Gumbel Probability Plot (Sheet 15) ─────────
# Same method as Figure 13, but n=9
# =============================================================================
d15 = to_num(
    read_sheet('15_GT_Prob_Plot'),
    [
        'Rank', 'Year', 'Observed (mm)',
        'Plot Position p', 'Gumbel yT', 'Fitted (mm)', 'Residual'
    ]
)
d15 = d15.dropna(subset=['Observed (mm)', 'Gumbel yT'])

print(f"\nSheet 15 | GT Daily Prob Plot n={len(d15)}")
print(" u=52.01 mm, β=14.38 mm | KS D=0.302, p=0.316 ✓")
print(" CAUTION: n=9 is very small → wide confidence intervals")

fig, axes = plt.subplots(1, 2, figsize=(15, 7))

ax = axes[0]
ax.scatter(
    d15['Gumbel yT'], d15['Observed (mm)'],
    color=B, s=80, zorder=3,
    label='GT observed (n=9)\n1923–1926, 1933, 1937–1940'
)
ax.plot(
    d15['Gumbel yT'], d15['Fitted (mm)'],
    color=R, lw=2.5,
    label='Gumbel EV1 fit\nu=52.01 mm, β=14.38 mm\nKS p=0.316 ✓'
)
for _, row in d15.iterrows():
    if pd.notna(row.get('Year')):
        ax.annotate(
            f"{int(row['Year'])}",
            xy=(row['Gumbel yT'], row['Observed (mm)']),
            xytext=(3, 3), textcoords='offset points',
            fontsize=7
        )
ax.set_xlabel('Gumbel Reduced Variate yT')
ax.set_ylabel('Annual Max Daily Rainfall (mm)')
ax.set_title(
    '(a) Gumbel Probability Plot – GT Daily (n=9)\n'
    'CAUTION: Small sample → wide CI',
    fontweight='bold'
)
ax.legend(fontsize=9)

ax2 = axes[1]
ax2.bar(
    d15['Year'].astype(int), d15['Residual'],
    color=[B if v >= 0 else R for v in d15['Residual']],
    alpha=0.75
)
ax2.axhline(0, color='black', lw=1)
ax2.set_xlabel('Year')
ax2.set_ylabel('Residual (mm)')
ax2.set_title(
    f'(b) Residuals | Mean = {d15["Residual"].mean():.2f} mm',
    fontweight='bold'
)
fig.suptitle(
    'Figure 14 – Gumbel EV1 Probability Plot: GT Daily | n=9',
    fontsize=12, fontweight='bold', y=1.01
)
savefig('Fig14_GT_ProbPlot')

# =============================================================================
# ── CELL 18: FIGURE 15 — Frequency Curves + ERA Table 5.19 (Sheet 16) ────────
# ERA values: ONLY from uploaded CSV (Table 5.19)
# Gumbel quantile: R₂₄(T) = u + β × [−ln(−ln(1−1/T))]
# =============================================================================
d16 = to_num(
    read_sheet('16_Frequency_Curves'),
    [
        'Return Period T (yr)', 'BOLE Gumbel (mm)',
        'GT Gumbel (mm)', 'NOAA Gumbel (mm)', 'BOLE GEV (mm)'
    ]
)
d16 = d16.dropna(subset=['Return Period T (yr)', 'BOLE Gumbel (mm)'])

print(
    f"\nSheet 16 | Frequency curves for T={d16['Return Period T (yr)'].values[:6]}"
)
print(" ERA Table 5.19:")
print(era.to_string(index=False))

fig, ax = plt.subplots(figsize=(12, 8))

ax.semilogx(
    d16['Return Period T (yr)'], d16['BOLE Gumbel (mm)'],
    color=G, lw=3,
    label='BOLE Gumbel EV1 [PRIMARY | u=39.61, β=14.03]'
)
ax.semilogx(
    d16['Return Period T (yr)'], d16['GT Gumbel (mm)'],
    color=B, lw=2.5, ls='--',
    label='GT Gumbel EV1 [n=9 | u=52.01, β=14.38]'
)
ax.semilogx(
    d16['Return Period T (yr)'], d16['NOAA Gumbel (mm)'],
    color=O, lw=2, ls='-.',
    label='NOAA 20CRv3 Gumbel [u=14.30, β=1.95] † NOT for design'
)
ax.semilogx(
    d16['Return Period T (yr)'], d16['BOLE GEV (mm)'],
    color=P, lw=2, ls=':',
    label='BOLE GEV (ξ=−0.158, Weibull type)'
)
ax.scatter(
    era['T_yr'], era['R24_mm'],
    color=R, s=90, marker='D', zorder=5,
    label='ERA Table 5.19'
)
ax.plot(
    era['T_yr'], era['R24_mm'],
    color=R, lw=2, ls='--', alpha=0.7
)
ax.set_xlabel('Return Period T (years)')
ax.set_ylabel('24-hr Design Rainfall R₂₄ (mm)')
ax.set_xticks([2, 5, 10, 25, 50, 100, 200, 500])
ax.set_xticklabels([2, 5, 10, 25, 50, 100, 200, 500])
ax.set_title(
    'Figure 15 – Gumbel EV1 Frequency Curves: All Sources + ERA Table 5.19\n'
    'R₂₄(T) = u + β×[−ln(−ln(1−1/T))]\n'
    '† NOAA severely underestimates extremes — DO NOT use for design',
    fontweight='bold'
)
ax.legend(fontsize=9)
savefig('Fig15_Frequency_Curves')

# =============================================================================
# ── CELL 19: FIGURE 16 — Design Depths vs ERA Table 5.19 (Sheets 17 + 18) ────
# ERA values: ONLY from uploaded CSV (Table 5.19)
# =============================================================================
d17 = to_num(
    read_sheet('17_Design_Depths'),
    [
        'T (yr)', 'BOLE Gumbel (mm)', 'GT Gumbel (mm)', 'NOAA (mm)',
        'CI Low (mm)', 'CI High (mm)'
    ]
)
d17 = d17.dropna(subset=['T (yr)', 'BOLE Gumbel (mm)'])
T6 = d17['T (yr)'].values.astype(float)
bole6 = d17['BOLE Gumbel (mm)'].values
gt6 = d17['GT Gumbel (mm)'].values
noaa6 = d17['NOAA (mm)'].values
ci_lo = d17['CI Low (mm)'].values
ci_hi = d17['CI High (mm)'].values
era_int = np.interp(T6, era['T_yr'].values, era['R24_mm'].values)

print("\nSheet 17 | Design depths at standard T values:")
print(f"{'T':>5} {'BOLE':>8} {'GT':>8} {'ERA':>9} {'NOAA':>8} {'BOLE-ERA':>10} {'%':>8}")
for i, T in enumerate(T6):
    print(
        f"{T:5.0f} {bole6[i]:8.1f} {gt6[i]:8.1f} {era_int[i]:9.2f} {noaa6[i]:8.1f} "
        f"{bole6[i]-era_int[i]:+10.1f} "
        f"{(bole6[i]-era_int[i])/era_int[i]*100:+7.1f}%"
    )

fig, axes = plt.subplots(1, 2, figsize=(16, 8))

ax = axes[0]
ax.plot(
    T6, bole6,
    color=G, lw=3, marker='o', ms=9,
    label='BOLE Gumbel EV1 [Primary, n=54]'
)
ax.fill_between(
    T6, ci_lo, ci_hi,
    color=G, alpha=0.2, label='BOLE 90% CI'
)
ax.plot(
    T6, gt6,
    color=B, lw=2.5, marker='s', ms=7, ls='--',
    label='GT Gumbel EV1 (n=9)'
)
ax.plot(
    T6, era_int,
    color=R, lw=2.5, marker='D', ms=8,
    label='ERA Table 5.19'
)
ax.plot(
    T6, noaa6,
    color=O, lw=1.5, marker='x', ms=8, ls=':',
    label='NOAA 20CRv3 † NOT for design'
)
ax.set_xscale('log')
ax.set_xlabel('Return Period T (years)')
ax.set_ylabel('24-hr Design Depth R₂₄ (mm)')
ax.set_xticks(T6.astype(int))
ax.set_xticklabels(T6.astype(int))
ax.set_title(
    '(a) 24-hr Design Depths\nERA from Table 5.19',
    fontweight='bold'
)
ax.legend(fontsize=8.5)

ax2 = axes[1]
bole_d = bole6 - era_int
gt_d = gt6 - era_int
noaa_d = noaa6 - era_int
ax2.plot(
    T6, bole_d,
    color=G, lw=3, marker='o', ms=7,
    label='BOLE − ERA T5.19'
)
ax2.plot(
    T6, gt_d,
    color=B, lw=2.5, marker='s', ms=6, ls='--',
    label='GT − ERA T5.19'
)
ax2.plot(
    T6, noaa_d,
    color=O, lw=1.5, marker='x', ms=7, ls=':',
    label='NOAA − ERA T5.19'
)
ax2.axhline(0, color='black', lw=1.5)
ax2.fill_between(
    T6, bole_d, 0, where=bole_d > 0,
    color=G, alpha=0.15
)
ax2.fill_between(
    T6, bole_d, 0, where=bole_d < 0,
    color=R, alpha=0.15
)
ax2.set_xscale('log')
ax2.set_xlabel('Return Period T (years)')
ax2.set_ylabel('Difference from ERA Table 5.19 (mm)')
ax2.set_xticks(T6.astype(int))
ax2.set_xticklabels(T6.astype(int))
for i, T_pt in enumerate(T6):
    ax2.annotate(
        f'Δ={bole_d[i]:+.1f}',
        xy=(T_pt, bole_d[i]),
        xytext=(0, 8), textcoords='offset points',
        ha='center', fontsize=8,
        color='darkgreen', fontweight='bold'
    )
ax2.set_title(
    '(b) Difference: BOLE/GT/NOAA vs ERA Table 5.19',
    fontweight='bold'
)
ax2.legend(fontsize=9)

fig.suptitle(
    f'Figure 16 – 24-hr Design Depths vs ERA Table 5.19\n'
    f'ERA: T2={era_int[0]:.2f}, T10={era_int[2]:.2f}, T100={era_int[5]:.2f} mm',
    fontsize=11, fontweight='bold', y=1.01
)
savefig('Fig16_Design_Depths_ERA519')

# =============================================================================
# ── CELL 20: FIGURE 17 — BOLE Gumbel 90% CI (Sheet 18) ───────────────────────
# Chow et al. (1988) variance formula:
# Var(xT) = (σ²/n) × [1 + 1.1396×yT + 1.1×yT²]
# 90% CI: xT ± 1.645 × √Var(xT)
# =============================================================================
d18 = to_num(
    read_sheet('18_Confidence_Intervals'),
    [
        'T (yr)', 'Best Estimate (mm)', 'CI Low 90% (mm)',
        'CI High 90% (mm)', 'CI Width (mm)', 'Rel Uncertainty (%)'
    ]
)
d18 = d18.dropna(subset=['T (yr)', 'Best Estimate (mm)'])

print(f"\nSheet 18 | {len(d18)} T values")
print(
    " Chow et al. (1988): Var(xT) = "
    "(σ²/n)×[1+1.1396×yT+1.1×yT²]"
)
print(" 90% CI: xT ± 1.645×√Var(xT) with σ=17.99 mm, n=54")

fig, axes = plt.subplots(1, 2, figsize=(15, 7))

ax = axes[0]
ax.plot(
    d18['T (yr)'], d18['Best Estimate (mm)'],
    color=G, lw=3, marker='o', ms=6,
    label='Best estimate (Gumbel EV1)'
)
ax.fill_between(
    d18['T (yr)'],
    d18['CI Low 90% (mm)'], d18['CI High 90% (mm)'],
    color=G, alpha=0.25, label='90% CI'
)
ax.scatter(
    era['T_yr'], era['R24_mm'],
    color=R, s=80, marker='D', zorder=5,
    label='ERA Table 5.19'
)
ax.set_xscale('log')
ax.set_xlabel('Return Period T (years)')
ax.set_ylabel('24-hr Design Depth (mm)')
ax.set_xticks([2, 5, 10, 25, 50, 100])
ax.set_xticklabels([2, 5, 10, 25, 50, 100])
ax.set_title(
    '(a) BOLE Gumbel EV1 with 90% CI\n'
    'Var(xT)=(σ²/n)×[1+1.1396yT+1.1yT²]',
    fontweight='bold'
)
ax.legend(fontsize=9)

ax2 = axes[1]
ax2.semilogx(
    d18['T (yr)'], d18['Rel Uncertainty (%)'],
    color=P, lw=2.5, marker='o', ms=5
)
ax2.axhline(
    20, color=R, lw=1.5, ls='--',
    alpha=0.7, label='20% reference'
)
ax2.set_xlabel('Return Period T (years)')
ax2.set_ylabel('Relative Uncertainty %')
ax2.set_xticks([2, 5, 10, 25, 50, 100, 200, 500])
ax2.set_xticklabels([2, 5, 10, 25, 50, 100, 200, 500])
ax2.set_title(
    '(b) Relative Uncertainty % vs T',
    fontweight='bold'
)
ax2.legend(fontsize=9)

fig.suptitle(
    'Figure 17 – BOLE Gumbel 90% Confidence Intervals | n=54',
    fontsize=12, fontweight='bold', y=1.01
)
savefig('Fig17_Confidence_Intervals')

# =============================================================================
# ── CELL 21: FIGURE 18 — Gumbel EV1 vs GEV (Sheet 19) ────────────────────────
# GEV fitted by MLE: shape ξ=−0.158 (negative → Weibull, bounded upper tail)
# =============================================================================
d19 = to_num(
    read_sheet('19_Gumbel_vs_GEV'),
    ['T (yr)', 'Gumbel EV1 (mm)', 'GEV (mm)',
     'Difference GEV−Gumbel (mm)']
)
d19 = d19.dropna(subset=['T (yr)', 'Gumbel EV1 (mm)'])

print("\nSheet 19 | Gumbel vs GEV comparison")
print(" GEV: ξ=−0.158, μ=39.7 mm, σ=11.3 mm (MLE, Jenkinson 1955)")
print(
    f" T=10: diff="
    f"{d19[d19['T (yr)']==10.0]['Difference GEV−Gumbel (mm)'].values[0]:.2f} mm"
)
print(
    f" T=100: diff="
    f"{d19[d19['T (yr)']==100.0]['Difference GEV−Gumbel (mm)'].values[0]:.2f} mm"
)
print(" All diffs within 90% CI → Gumbel EV1 retained as primary model")

fig, axes = plt.subplots(1, 2, figsize=(15, 7))

ax = axes[0]
ax.plot(
    d19['T (yr)'], d19['Gumbel EV1 (mm)'],
    color=G, lw=3, marker='o', ms=5,
    label='Gumbel EV1 [Adopted primary model]'
)
ax.plot(
    d19['T (yr)'], d19['GEV (mm)'],
    color=P, lw=2.5, ls='--', marker='s', ms=5,
    label='GEV (ξ=−0.158, Weibull type — bounded upper tail)'
)
ax.scatter(
    era['T_yr'], era['R24_mm'],
    color=R, s=80, marker='D', zorder=5,
    label='ERA Table 5.19'
)
ax.set_xscale('log')
ax.set_xlabel('Return Period T (years)')
ax.set_ylabel('24-hr Design Depth (mm)')
ax.set_xticks([2, 5, 10, 25, 50, 100, 200, 500])
ax.set_xticklabels([2, 5, 10, 25, 50, 100, 200, 500])
ax.set_title(
    '(a) Gumbel EV1 vs GEV Frequency Curves',
    fontweight='bold'
)
ax.legend(fontsize=9)

ax2 = axes[1]
diff = d19['Difference GEV−Gumbel (mm)']
ax2.semilogx(
    d19['T (yr)'], diff,
    color=P, lw=2.5, marker='s', ms=6
)
ax2.axhline(0, color='black', lw=1)
ax2.fill_between(
    d19['T (yr)'], diff, 0, where=diff > 0,
    color=P, alpha=0.2, label='GEV > Gumbel'
)
ax2.fill_between(
    d19['T (yr)'], diff, 0, where=diff < 0,
    color=G, alpha=0.2, label='Gumbel > GEV'
)
ax2.set_xlabel('Return Period T (years)')
ax2.set_ylabel('Difference: GEV − Gumbel EV1 (mm)')
ax2.set_xticks([2, 5, 10, 25, 50, 100, 200, 500])
ax2.set_xticklabels([2, 5, 10, 25, 50, 100, 200, 500])
ax2.set_title(
    '(b) Difference GEV − Gumbel\n'
    'T=10: −1.0 mm | T=100: +5.7 mm → within 90% CI',
    fontweight='bold'
)
ax2.legend(fontsize=9)

fig.suptitle(
    'Figure 18 – Gumbel EV1 vs GEV: BOLE Station | GEV shape ξ=−0.158',
    fontsize=12, fontweight='bold', y=1.01
)
savefig('Fig18_Gumbel_vs_GEV')

# =============================================================================
# ── CELL 22: FIGURE 19 — Annual Relative Bias % (Sheet 20) ───────────────────
# Relative Bias = (NOAA − GT) / GT × 100
# =============================================================================
d20 = to_num(
    read_sheet('20_Annual_Rel_Bias'),
    ['Year', 'GT (mm)', 'NOAA (mm)',
     'Abs Bias (mm)', 'Rel Bias (%)']
)
d20 = d20.dropna(subset=['Year', 'Rel Bias (%)'])
mean_rb = d20['Rel Bias (%)'].mean()

print(
    f"\nSheet 20 | Mean Rel Bias = +{mean_rb:.2f}% | "
    f"n_pos={(d20['Rel Bias (%)']>0).sum()} | "
    f"n_neg={(d20['Rel Bias (%)']<0).sum()}"
)

fig, ax = plt.subplots(figsize=(14, 6))
ax.bar(
    d20['Year'], d20['Rel Bias (%)'],
    color=[R if v > 0 else B for v in d20['Rel Bias (%)']],
    alpha=0.8, width=0.9
)
ax.axhline(0, color='black', lw=0.8)
ax.axhline(
    mean_rb, color=R, lw=2.5, ls='--',
    label=f'Mean = +{mean_rb:.1f}%'
)
ax.set_xlabel('Year')
ax.set_ylabel('Relative Bias % (NOAA−GT)/GT×100')
ax.set_title(
    'Figure 19 – Annual Relative Bias %: NOAA vs GT\n'
    'Positive (red) = NOAA overestimates | Negative (blue) = underestimates\n'
    f'n_positive={(d20["Rel Bias (%)"]>0).sum()}, '
    f'n_negative={(d20["Rel Bias (%)"]<0).sum()} | Mean=+{mean_rb:.1f}%',
    fontweight='bold'
)
ax.legend(fontsize=10)
savefig('Fig19_Annual_Rel_Bias')

# =============================================================================
# ── CELL 23: FIGURE 20 — Seasonal Monthly Bias % (Sheet 21) ──────────────────
# =============================================================================
d21 = to_num(
    read_sheet('21_Seasonal_Bias'),
    ['GT Mean (mm)', 'NOAA Mean (mm)', 'Bias (mm)', 'Bias (%)']
)
d21 = d21.dropna(subset=['GT Mean (mm)'])
months21 = [
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'
]
d21['Month'] = months21[:len(d21)]
season_map = {
    'Jan': 'Dry',  'Feb': 'Dry',
    'Mar': 'Belg', 'Apr': 'Belg', 'May': 'Belg',
    'Jun': 'Kiremt', 'Jul': 'Kiremt', 'Aug': 'Kiremt', 'Sep': 'Kiremt',
    'Oct': 'Dry', 'Nov': 'Dry', 'Dec': 'Dry'
}
d21['Season'] = [season_map[m] for m in d21['Month']]
sc_map = {'Dry': GR, 'Belg': O, 'Kiremt': B}

print("\nSheet 21 | Monthly relative bias:")
for _, row in d21.iterrows():
    print(
        f" {row['Month']:3s}: "
        f"GT={row['GT Mean (mm)']:6.1f} "
        f"NOAA={row['NOAA Mean (mm)']:6.1f} "
        f"Bias={row['Bias (%)']:+5.1f}% "
        f"[{row['Season']}]"
    )

fig, ax = plt.subplots(figsize=(12, 7))
ax.bar(
    np.arange(len(d21)), d21['Bias (%)'],
    color=[sc_map[s] for s in d21['Season']],
    alpha=0.85, edgecolor='white'
)
ax.axhline(0, color='black', lw=0.8)
ax.axhline(
    d21['Bias (%)'].mean(),
    color=R, lw=2, ls='--',
    label=f'Overall mean = +{d21["Bias (%)"].mean():.1f}%'
)
ax.set_xticks(np.arange(len(d21)))
ax.set_xticklabels(d21['Month'])
ax.set_xlabel('Month')
ax.set_ylabel('Monthly Relative Bias % (NOAA − GT) / GT × 100')
for i, (_, row) in enumerate(d21.iterrows()):
    ax.text(
        i, row['Bias (%)'] + 0.5,
        f"{row['Bias (%)']:.1f}%",
        ha='center', fontsize=8, fontweight='bold'
    )
ax.set_title(
    'Figure 20 – Seasonal Monthly Relative Bias %: NOAA vs GT\n'
    'Belg = Mar–May | Kiremt = Jun–Sep | Dry = Oct–Feb',
    fontweight='bold'
)
ax.legend(
    handles=[
        mpatches.Patch(color=c, alpha=0.85, label=l)
        for l, c in sc_map.items()
    ] + ax.lines[:1],
    fontsize=9
)
savefig('Fig20_Seasonal_Bias')

# =============================================================================
# ── CELL 24: FIGURE 21 — Annual Max Daily Bias (Sheet 22) ────────────────────
# 9 overlap years: 1923–1926, 1933, 1937–1940
# =============================================================================
d22 = to_num(
    read_sheet('22_Daily_Max_Bias'),
    [
        'Year', 'NOAA Max Daily (mm)', 'GT Max Daily (mm)',
        'Abs Bias (mm)', 'Rel Bias (%)', 'GT/NOAA Ratio'
    ]
)
d22 = d22.dropna(subset=['Year', 'GT Max Daily (mm)'])

print(f"\nSheet 22 | n={len(d22)} overlap years")
print(
    f" Mean GT/NOAA ratio = {d22['GT/NOAA Ratio'].mean():.2f}× | "
    f"Mean rel bias = {d22['Rel Bias (%)'].mean():.1f}%"
)
print(
    f" NOTE: thesis states ~4.5× — data gives "
    f"{d22['GT/NOAA Ratio'].mean():.2f}×. Please verify."
)

fig, axes = plt.subplots(2, 1, figsize=(12, 11))

ax = axes[0]
xd22 = np.arange(len(d22))
ax.bar(
    xd22 - 0.2, d22['GT Max Daily (mm)'], 0.38,
    color=B, alpha=0.85,
    label=f'GT (mean={d22["GT Max Daily (mm)"].mean():.1f} mm)'
)
ax.bar(
    xd22 + 0.2, d22['NOAA Max Daily (mm)'], 0.38,
    color=O, alpha=0.85,
    label=f'NOAA (mean={d22["NOAA Max Daily (mm)"].mean():.1f} mm)'
)
ax.set_xticks(xd22)
ax.set_xticklabels(d22['Year'].astype(int), rotation=45)
ax.set_ylabel('Annual Max Daily Rainfall (mm)')
ax.set_title(
    '(a) GT vs NOAA Annual Max Daily (9 Overlap Years)',
    fontweight='bold'
)
ax.legend(fontsize=9)

ax2 = axes[1]
ax2b = ax2.twinx()
ax2.bar(
    xd22, d22['Rel Bias (%)'],
    color=R, alpha=0.65, width=0.6,
    label='Rel Bias %'
)
ax2.axhline(
    d22['Rel Bias (%)'].mean(),
    color=R, lw=2, ls='--',
    label=f'Mean = {d22["Rel Bias (%)"].mean():.1f}%'
)
ax2b.plot(
    xd22, d22['GT/NOAA Ratio'],
    color=B, lw=2.5, marker='^', ms=7,
    label=f'GT/NOAA ratio (mean={d22["GT/NOAA Ratio"].mean():.2f}×)'
)
ax2.set_xticks(xd22)
ax2.set_xticklabels(d22['Year'].astype(int), rotation=45)
ax2.set_ylabel('Relative Bias %', color=R)
ax2b.set_ylabel('GT / NOAA Ratio', color=B)
ax2.tick_params(axis='y', labelcolor=R)
ax2b.tick_params(axis='y', labelcolor=B)
ax2.set_title(
    f'(b) Rel Bias % and GT/NOAA Ratio\n'
    f'NOAA underestimates by mean '
    f'{abs(d22["Rel Bias (%)"].mean()):.1f}% | '
    f'ratio={d22["GT/NOAA Ratio"].mean():.2f}×',
    fontweight='bold'
)
l1, lb1 = ax2.get_legend_handles_labels()
l2, lb2 = ax2b.get_legend_handles_labels()
ax2.legend(l1 + l2, lb1 + lb2, fontsize=8)

fig.suptitle(
    'Figure 21 – Annual Max Daily Bias: NOAA vs GT\n9 overlap years',
    fontsize=12, fontweight='bold', y=1.01
)
savefig('Fig21_Daily_Max_Bias')

# =============================================================================
# ── CELL 25: FIGURE 22 — Variance Structure (Sheet 23) ───────────────────────
# OLS slope < 1 indicates NOAA compresses inter-annual variability
# CV comparison: GT CV=0.205 vs NOAA CV=0.148
# =============================================================================
d23 = to_num(
    read_sheet('23_Variance_Structure'),
    ['Year', 'GT (mm)', 'NOAA (mm)', 'OLS Predicted (mm)']
)
d23 = d23.dropna(subset=['GT (mm)', 'NOAA (mm)'])
sl23, ic23, r23, p23, _ = stats.linregress(
    d23['NOAA (mm)'], d23['GT (mm)']
)
xs23 = np.linspace(
    d23['NOAA (mm)'].min(), d23['NOAA (mm)'].max(), 200
)
cv_gt = d23['GT (mm)'].std(ddof=1) / d23['GT (mm)'].mean()
cv_noaa = d23['NOAA (mm)'].std(ddof=1) / d23['NOAA (mm)'].mean()

print(
    f"\nSheet 23 | OLS slope={sl23:.4f} | r={r23:.3f} | p={p23:.4f}"
)
print(
    f" GT CV={cv_gt:.3f} | NOAA CV={cv_noaa:.3f} | "
    f"Ratio={cv_gt/cv_noaa:.2f}"
)
print(
    f" Conclusion: NOAA compresses inter-annual variability by "
    f"{(1-cv_noaa/cv_gt)*100:.0f}%"
)

fig, axes = plt.subplots(1, 2, figsize=(15, 7))

ax = axes[0]
sc23 = ax.scatter(
    d23['NOAA (mm)'], d23['GT (mm)'],
    c=d23['Year'], cmap='viridis', s=55,
    alpha=0.85, zorder=3
)
plt.colorbar(sc23, ax=ax, label='Year')
ll23 = [
    min(d23['NOAA (mm)'].min(), d23['GT (mm)'].min()) - 100,
    max(d23['NOAA (mm)'].max(), d23['GT (mm)'].max()) + 100
]
ax.plot(
    ll23, ll23,
    'k--', lw=1.5, alpha=0.6, label='1:1 line'
)
ax.plot(
    xs23, sl23 * xs23 + ic23,
    color=R, lw=2.5,
    label=f'OLS slope={sl23:.4f}\nr={r23:.3f}, p={p23:.4f}'
)
ax.set_xlabel('NOAA 20CRv3 Annual Total (mm)')
ax.set_ylabel('GT Annual Total (mm)')
ax.set_title(
    f'(a) Variance Structure Scatter\n'
    f'GT CV={cv_gt:.3f} vs NOAA CV={cv_noaa:.3f}',
    fontweight='bold'
)
ax.legend(fontsize=9)

ax2 = axes[1]
ax2.plot(
    d23['Year'], d23['GT (mm)'],
    color=B, lw=2,
    label=f'GT (SD={d23["GT (mm)"].std(ddof=1):.1f} mm, CV={cv_gt:.3f})'
)
ax2.plot(
    d23['Year'], d23['NOAA (mm)'],
    color=O, lw=2, ls='--',
    label=f'NOAA (SD={d23["NOAA (mm)"].std(ddof=1):.1f} mm, CV={cv_noaa:.3f})'
)
ax2.set_xlabel('Year')
ax2.set_ylabel('Annual Total (mm)')
ax2.set_title(
    '(b) Time Series: NOAA Swings Smaller than GT\n'
    'NOAA suppresses variability',
    fontweight='bold'
)
ax2.legend(fontsize=9)

fig.suptitle(
    'Figure 22 – Variance Structure: NOAA vs GT\nOLS slope=1.18; CV ratio=1.38',
    fontsize=12, fontweight='bold', y=1.01
)
savefig('Fig22_Variance_Structure')

# =============================================================================
# ── CELL 26: Summary print ───────────────────────────────────────────────────
# =============================================================================
import glob
figs = sorted(glob.glob(os.path.join(FIG_DIR, '*.png')))
print("\n" + "="*60)
print(f"✓ All {len(figs)} figures saved to folder: '{FIG_DIR}/'")
print("="*60)
for f in figs:
    size_kb = os.path.getsize(f) / 1024
    print(f" {os.path.basename(f):45s} {size_kb:6.0f} KB")

print("\nKEY CORRECTIONS APPLIED IN THIS SCRIPT:")
print(" 1. Pettitt K=168, p=0.2145")
print(" 2. ERA values = Table 5.19 from uploaded CSV only")
```
